```python
# Data PreProcessing
```

```python
for x in df.columns:
    if df[x].dtype=='object':
        df[x]=df[x].astype('category')
```

```python
df['day']=df['day'].astype('category')
```

```python
X=df.drop('Target',axis=1)
```

```python
y=df['Target']
y=y.replace({'yes':1,'no':0})
```

```python
X=pd.get_dummies(X,drop_first=True)
X.head()
```

```python
#X['pdays']=X.pdays.replace(-1,0)
```

```python
X.head()
```

```python
X.info()
```

```python
# Feature Scaling for some numeric columns which have great variation in their range
# balance,duration,campaign,pdays,previous
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scale_df=X.copy()
cols=['balance','campaign','pdays','previous']
features=scale_df[cols]
sca=scaler.fit(features.values)
features=sca.transform(features.values)
scale_df[cols]=features
print(scale_df.head())
```

```python
# Data Modelling
# Split into Train and Test data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
## Create model with class_weight parameter to account for Imbalanced class outputs
logistic_model=LogisticRegression(class_weight='balanced')
X_train,X_test,y_train,y_test=train_test_split(scale_df,y,test_size=0.30,random_state=49)
logistic_model.fit(X_train,y_train)
```

```python
train_score=logistic_model.score(X_train,y_train)
test_score=logistic_model.score(X_test,y_test)
print("Train Model Score : {}".format(train_score))
print("Test Model Score : {}".format(test_score))
```

```python
y_pred=logistic_model.predict(X_test)
```

```python
## function to get confusion matrix in a proper format
def draw_cm( actual, predicted ):
    cm = confusion_matrix( actual, predicted)
    sns.heatmap(cm, annot=True, fmt='d',cmap='Blues',xticklabels = [0,1] , yticklabels = [0,1] )
    plt.ylabel('Observed')
    plt.xlabel('Predicted')
    plt.show()
```

```python
## Confusion Matrix 
from sklearn.metrics import confusion_matrix,recall_score,precision_score,f1_score,roc_auc_score

draw_cm(y_test,y_pred)
```

```python
print("Recall: {}".format(recall_score(y_test,y_pred)))

print("Precision: {}".format(precision_score(y_test,y_pred)))

print("f1 score: {}".format(f1_score(y_test,y_pred)))

print("ROC_AUC_Score: {}".format(roc_auc_score(y_test,y_pred)))
```

```python
## roc curve
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logistic_model.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logistic_model.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")

plt.show()
```

```python
# Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=1)
dtree.fit(X_train,y_train)
print("Train Score: {}".format(dtree.score(X_train,y_train)))
print("Test Score: {}".format(dtree.score(X_test,y_test)))
```

```python
y_pred=dtree.predict(X_test)
draw_cm(y_test,y_pred)
```

```python
# Feature Importances 
pd.DataFrame(dtree.feature_importances_,columns=['Imp'],index=X_train.columns).sort_values(by='Imp',ascending=False)
```
